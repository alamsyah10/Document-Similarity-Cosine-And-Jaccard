```
import numpy as np
import math
import re
import string
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```
#load data
data = []
data.append(open('1.txt').read())
data.append(open('2.txt').read())
data.append(open('3.txt').read())
data.append(open('4.txt').read())
data.append(open('5.txt').read())
data.append(open('6.txt').read())
data.append(open('7.txt').read())
data.append(open('8.txt').read())
data.append(open('9.txt').read())
data.append(open('10.txt').read())

for i in data:
  print(i)
  print()
```

```
#text preprocessing
token = []
count = 0
for i in data:
  print(i)
  print()
  abstract = i.translate(str.maketrans('','',string.punctuation)).lower()
  abstract = re.sub(r"\d+", "", abstract)
  temp_token = nltk.tokenize.word_tokenize(abstract)

  listStopword =  set(stopwords.words('english'))
  
  removed = []
  for t in temp_token:
      if t not in listStopword:
          removed.append(t)

  stemmer = SnowballStemmer('english') 

  sb_stemming = []
  for k in removed:
    sb_stemming.append(stemmer.stem(k))

  token.append(sb_stemming)

  print(token[count])
  print()

  count += 1
```

```
#representasi fitur vectorized

dct = {}

tokens = np.array(token) 

for i in range(tokens.size):
      tempToken = np.array(tokens[i])
      for j in range(tempToken.size):
            dct[tempToken[j]] = 0

arrayOfWord = []

for i in range(tokens.size):
      tempToken = np.array(tokens[i])
      for j in range(tempToken.size):
            if dct[tempToken[j]] == 0:
                dct[tempToken[j]] = 1
                arrayOfWord.append(tempToken[j])

vectorizeds = []
arrayOfWord = np.array(arrayOfWord)
tokens = np.array(token) 

indexOfArrayWord = {}
idx = 0

for i in arrayOfWord:
    indexOfArrayWord[i] = idx
    idx = idx+1
  
for i in range(tokens.size):
      vectorized = []
      for j in arrayOfWord:
            vectorized.append({j : 0})
      vectorizeds.append(vectorized)

for i in range(tokens.size):
      tempToken = np.array(tokens[i])
      for j in range(tempToken.size):
            vectorizeds[i][indexOfArrayWord[tempToken[j]]][tempToken[j]] = vectorizeds[i][indexOfArrayWord[tempToken[j]]][tempToken[j]]+1

vectorizeds = np.array(vectorizeds)

print(vectorizeds)
```

```
#Jaccard Similarity

def jaccard(vecDoc1, vecDoc2):
    numOfSlice = 0
    totalWord = 0
    for i in range(vecDoc1.size):
        x1 = vecDoc1[i][arrayOfWord[i]]
        x2 = vecDoc2[i][arrayOfWord[i]]

        if ((x1 > 0) or (x2 > 0)):
            totalWord = totalWord + 1
        if ((x1 > 0) and (x2 > 0)):
            numOfSlice = numOfSlice + 1

    return numOfSlice/totalWord * 100

jaccards = []
tempJaccards = []

for i in range(int(int(vectorizeds.size/arrayOfWord.size))):
    tempJaccards = []
    for j in range(int(int(vectorizeds.size/arrayOfWord.size))):
        temp = jaccard(vectorizeds[i], vectorizeds[j])
        tempJaccards.append(temp)
    jaccards.append(tempJaccards)

jaccards = np.array(jaccards)

print(jaccards)
```
