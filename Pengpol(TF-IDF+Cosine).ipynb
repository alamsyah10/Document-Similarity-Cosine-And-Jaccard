```
import numpy as np
import math
import re
import string
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
```

```
#load data
data = []
data.append(open('1.txt').read())
data.append(open('2.txt').read())
data.append(open('3.txt').read())
data.append(open('4.txt').read())
data.append(open('5.txt').read())
data.append(open('6.txt').read())
data.append(open('7.txt').read())
data.append(open('8.txt').read())
data.append(open('9.txt').read())
data.append(open('10.txt').read())

for i in data:
  print(i)
  print()
```

```
#text preprocessing
token = []
count = 0
for i in data:
  print(i)
  print()
  abstract = i.translate(str.maketrans('','',string.punctuation)).lower()
  abstract = re.sub(r"\d+", "", abstract)
  temp_token = nltk.tokenize.word_tokenize(abstract)

  listStopword =  set(stopwords.words('english'))
  
  removed = []
  for t in temp_token:
      if t not in listStopword:
          removed.append(t)

  stemmer = SnowballStemmer('english') 

  sb_stemming = []
  for k in removed:
    sb_stemming.append(stemmer.stem(k))

  token.append(sb_stemming)

  print(token[count])
  print()

  count += 1
```

```
#TF
def tf(word, token):
  f = 0
  for n in token:
    if n == word:
      f += 1

  return f

TF = []

for i in token:
  temp_tf = {}
  for j in i:
    temp_tf[j] = tf(j, i)
  
  TF.append(temp_tf)

for i in TF:
  print(i)
```

```
#IDF
idfs = {}
dct = {}

tokens = np.array(token) 

for i in range(tokens.size):
      tempToken = np.array(tokens[i])
      for j in range(tempToken.size):
            dct[tempToken[j]] = 0

arrayOfWord = []

for i in range(tokens.size):
      tempToken = np.array(tokens[i])
      for j in range(tempToken.size):
            if dct[tempToken[j]] == 0:
                dct[tempToken[j]] = 1
                arrayOfWord.append(tempToken[j])

def idf(word, token):
    freq = 0
    for i in range(token.size):
        tempToken = np.array(token[i])
        for j in range(tempToken.size):
            if word == tempToken[j]:
                freq = freq + 1
                break
    
    return token.size/freq


arrayOfWord = np.array(arrayOfWord)

for i in arrayOfWord:
    tempIdf = idf(i, tokens)
    idfs[i] = tempIdf    

for i in idfs:
  print(i, ' ', idfs[i])
```

```
#TF-IDF
TF_IDF = []

for i in range(len(token)):
  temp_tfidf = {}
  for j in token[i]:
    temp_tfidf[j] = TF[i][j] * idfs[j]
  
  TF_IDF.append(temp_tfidf)

for i in TF_IDF:
  print(i)
```

```
#cosine similarity in percent

def cosSim(index1, index2):

  #get unique words in doc1 union doc2
  dct = {}
  words12 = []
  for i in token[index1]:
    for j in token[index2]:
      if i == j:
        dct[i] = 1
        continue

  for i in dct:
    words12.append(i)

  sum12 = 0
  for i in words12:
    sum12 += TF_IDF[index1][i] * TF_IDF[index2][i]

  #get unique words in doc1
  dct = {}
  words1 = []
  for i in token[index1]:
    dct[i] = 1

  for i in dct:
    words1.append(i)
  
  #get unique words in doc2
  dct = {}
  words2 = []
  for i in token[index2]:
    dct[i] = 1

  for i in dct:
    words2.append(i)

  #count cosine similarity
  p1 = 0
  for i in words1:
    p1 += TF_IDF[index1][i] * TF_IDF[index1][i]
  p1 = math.sqrt(p1)

  p2 = 0
  for i in words2:
    p2 += TF_IDF[index2][i] * TF_IDF[index2][i]
  p2 = math.sqrt(p2)
  
  return (sum12 / (p1 * p2))*100

cos_sim = []
for i in range(len(token)):
  cos_sim.append([])
  for j in range(len(token)):
    cos_sim[i].append(cosSim(i, j))

for i in cos_sim:
  print(i)
```
